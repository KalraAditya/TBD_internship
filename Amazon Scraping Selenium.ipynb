```python
links = ['https://www.amazon.in/Parle-Hide-Seek-Chocolate-Cookies/dp/B00TX97G24/ref=zg_bs_4861746031_sccl_1/257-8129200-4984438?psc=1',
'https://www.amazon.in/Unibic-UNIBIC-CHOCO-CHIP-COOKIES500G/dp/B07DLYK78Z/ref=zg_bs_4861746031_sccl_3/257-8129200-4984438?th=1',
'https://www.amazon.in/Sunfeast-Dark-Fantasy-Crunchy-Chocolatey/dp/B08PK85Z3X/ref=zg_bs_4861746031_sccl_5/257-8129200-4984438?psc=1',
'https://www.amazon.in/RiteBite-Max-Protein-Breakfast-Cookies/dp/B08FKGTPLJ/ref=zg_bs_4861746031_sccl_9/257-8129200-4984438?th=1',
'https://www.amazon.in/Ultimate-Chocochip-cookies-Premium-Gourmet/dp/B0BBMPCBMQ/ref=zg_bs_4861746031_sccl_20/257-8129200-4984438?psc=1',
'https://www.amazon.in/Wheafree-Gluten-Free-Chocochip-Cookies/dp/B0B777WH5Q/ref=zg_bs_4861746031_sccl_27/257-8129200-4984438?psc=1',
'https://www.amazon.in/dp/B093JG2CHN/ref=syn_sd_onsite_desktop_0?ie=UTF8&pd_rd_plhdr=t&th=1' ]
```

```python
rank1=[]
rank2=[]
for link in links:
    driver = webdriver.Chrome(r"C:\Users\tbd_d\OneDrive\Desktop\chromedriver.exe")
    driver.get(link)
    userid_element = driver.find_elements(By.XPATH,r"/html/body/div[2]/div[2]/div[5]/div[14]/div/div/div/div[2]/div[1]/div/div/table/tbody/tr[3]/td/span/span[1]")[0]
    userid = userid_element.text
    rank1.append(userid)
    userid_element2 = driver.find_elements(By.XPATH,r"/html/body/div[2]/div[2]/div[5]/div[14]/div/div/div/div[2]/div[1]/div/div/table/tbody/tr[3]/td/span/span[2]")[0]
    userid2 = userid_element2.text
    rank2.append(userid2)
    time.sleep(2)
    
```

```python
name = ['Parle Hide and seek','Unibic Choco Chip','Dark Fantasy','RiteBite Max Protein','Orion Ultimate ChocoChip Cookies','Wheafree Glutten-Free ChocoChip','TBD Dark Chocolate Cookies']
```

```python
df = pd.DataFrame()
```

```python
# Create a Timestamp for your output to track when data was collected

from datetime import datetime
date = datetime.now()
date = date.strftime("%d/%m/%Y %H:%M:%S")
date
```

```python
df.insert(loc=0, column='Date', value=[date]*7)
df.insert(loc=1, column='Item_Name', value=name)
df.insert(loc=2, column='Rank_C1', value=rank1)
df.insert(loc=3, column='Rank_C2', value=rank2)
```

```python
c1 = df['Rank_C1'].str.split("in", expand= True,)
c2 = df['Rank_C2'].str.split("in", expand= True,)

df['Rank_C1'] = c1[0]
df['C1'] = c1[1]
df['Rank_C2'] = c2[0]
df['C2'] = c2[1]

df['Rank_C1'] = df['Rank_C1'].str.replace('\W', '', regex=True)
df['Rank_C2'] = df['Rank_C2'].str.replace('\W', '', regex=True)
```

```python
df = df.append(df1,ignore_index=True)
```
